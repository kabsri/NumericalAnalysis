```python
def f(a, x):
    if x>0:
        return x**a
    return 0

def bucket(x, n):
    for k in range(1, n+1):
        if (-1+k*2/n) > x:
            return -1+(k*2-1)/n
    return 1

def pwConst(n, a, x):
    return f(a, bucket(x,n))

def maxNorm(xs, ys):
    return np.max(np.abs(np.array(xs)-np.array(ys)))

def pwConstNs(a, ns):
    data = []
    dx = np.linspace(-1,1,200)
    data.append(dx)
    y = []
    for x in dx:
        y.append(f(a, x))
    data.append(y)
    for n in ns:
        dy = []
        for x in dx:
            dy.append(pwConst(n, a, x))
        data.append(dy)
    return data

def pwConstTest(aList, ns):
    rs = []
    for a in aList:
        data = np.array(pwConstNs(a, ns))
        maxima = []
        for i in range(2, len(data)):
            maxima.append(maxNorm(data[1], data[i]))
        fit = np.polyfit(np.log(ns),np.log(maxima), deg=1)
        rs.append(-fit[0])
    return rs
```

```python
data = pwConstNs(0.1, [3,5,15])
plt.plot(data[0], data[1], color="orange")
plt.plot(data[0], data[2], color="green")
plt.plot(data[0], data[3], color="red")
plt.plot(data[0], data[4], color="blue")
plt.show()
```

```python
ns = np.array(range(50,300,7))
data = np.array(pwConstNs(2, ns))
maxima = []
for i in range(2, len(data)):
    maxima.append(maxNorm(data[1], data[i]))

fit = np.polyfit(np.log(ns),np.log(maxima), deg=1)
xs = np.linspace(50,300,100)
ys = np.exp(fit[1])*xs**fit[0]
plt.scatter(ns, maxima, color="blue")
plt.plot(xs, ys, color="red")
plt.show()
```
